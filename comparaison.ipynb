```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns   
```

```python
df = pd.concat([
    pd.read_csv('data/model_result_linear_resgression.csv'),
    pd.read_csv('data/model_result_mlpregressor.csv'),
    pd.read_csv('data/model_result_random_forest.csv')
], ignore_index=True)
df
```

```python
_, (ax1, ax2, ax3, ax4) = plt.subplots(1,4,figsize=(20, 5))
sns.barplot(df, y='MAE', hue='Model', ax=ax1)
ax1.set_title('MAE')
sns.barplot(df, y='MSE', hue='Model', ax=ax2)
ax2.set_title('MSE')
sns.barplot(df, y='RMSE', hue='Model', ax=ax3)
ax3.set_title('RMSE')
sns.barplot(df, y='R²', hue='Model', ax=ax4)
ax4.set_title('R²');
```

### Comparaison des modèles de régression

En comparant les trois modèles (**Linear Regression**, **MLPRegressor** et **Random Forest**) sur les métriques MAE, MSE, RMSE et R², Random Forest se distingue comme le meilleur choix. Il affiche les valeurs les plus basses pour MAE (~13000), MSE (~5) et RMSE (~2000), indiquant une meilleure précision et une réduction des erreurs par rapport à Linear Regression et MLPRegressor, qui ont des scores significativement plus élevés (par exemple, MAE de ~14000 pour MLPRegressor). De plus, Random Forest atteint un R² d’environ 0.8, proche de celui de MLPRegressor, mais avec des erreurs bien moindres, ce qui montre qu’il explique mieux la variance tout en étant plus robuste. Linear Regression, avec un R² plus faible (~0.75) et des erreurs plus élevées, est le moins performant. Ainsi, Random Forest est le modèle optimal pour ce cas.
