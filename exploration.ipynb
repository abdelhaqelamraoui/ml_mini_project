# After Cleaning

```python
df = pd.read_csv('data/data_cleaned.csv');
df.sample(5)
```

### Prix

```python
print(df['prix'].sort_values(ascending=True).head(10).to_frame())
print(df['prix'].sort_values(ascending=False).head(10).to_frame())
```

```python
_, (ax1, ax2) = plt.subplots(1, 2 , figsize=(14, 6))
sns.boxplot(df, y='prix', ax=ax1);
# sns.histplot(df, y='prix', bins=20, kde=True, ax=ax2);
sns.kdeplot(df, x='prix', ax=ax2);
# sns.violinplot(df, y='prix', ax=ax2);
```

```python
sns.kdeplot(df, x='prix');
# sns.kdeplot(np.log1p(df['prix']));
```

```python
tmp_df = df.copy()

q1 = tmp_df['prix'].quantile(0.25)
q3 = tmp_df['prix'].quantile(0.75)
iqr = q3 - q1


lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr


tmp_df_clean = tmp_df[(tmp_df['prix'] >= lower_bound) & (tmp_df['prix'] <= upper_bound)]
```

```python
sns.kdeplot(tmp_df_clean, x='prix')
# sns.kdeplot(np.log1p(df['prix']))
```

```python
tmp_df_clean['prix'].describe()
```
